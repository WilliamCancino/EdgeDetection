<div style="text-align: justify"><font size="4">
Edge detection is one of the fundamental operations when we perform image processing. It helps us reduce the amount of data represented by pixels to process and maintains the structural aspect of the image. We're going to look into three commonly used edge detection schemes, the gradient (Sobel) based edge detector, the Laplacian based edge detector and the Canny Edge Detector. The first two of them work with convolutions and the third is a various tools combination to get better results, so let's go on it.
</font></div>
<br><br><br>

### SOBEL DETECTION

<div style="text-align: justify"><font size="4">
Sobel edge detector is a gradient based method based on the first order derivatives. It calculates the first derivatives of the image separately for the X and Y axes. The operator uses two 3X3 kernels which are convolved with the original image to calculate approximations of the derivatives - one for horizontal changes, and one for vertical. The picture below shows Sobel Kernels in x-dir and y-dir:

Now that you have seen some theory, please go to the code example below.

```python
#FIRST WE IMPORT THE LIBRARIES THAT WE ARE GOING TO USE
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#THEN WE READ THE IMAGE FOR THE DETECTION,
#AS WE DON'T NEED FULL-COLOR FOR THIS,
#WE FIRST CONVERT THE IMAGE TO THE GRAY SCALE.

img0 = cv2.imread('img/mustang.jpg',) #reading the image into a variable.

gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY) # converting the image to the gray scale into another variable

#NOW WE NEED TO AVOID THE NOISE, SO IT'S NECESARY TO PASS A GAUSSIAN FILTER TO THE IMAGE

img = cv2.GaussianBlur(gray,(3,3),0) #filtering the image


#IT'S TIME TO CALL THE SOBEL METHODS FROM OpenCV

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x

sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y


#AT LAST WE JUST PLOT THE RESULTS WE ACHIEVED, USING Pyplot FROM Matplotlib

plt.figure(figsize=(10,6)) #determining the ploting size
plt.subplot(1,1,1),plt.imshow(img,cmap = 'gray') #Original image in gray scale
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(sobelx,cmap = 'gray') #Sobel in "x" direction
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(sobely,cmap = 'gray') #Sobel in "x" direction
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#image reading
img1 = cv2.imread('img/mustang.jpg',)
#img3 = cv2.imread('img/sudoku.jpg',) #you can try this example too, just uncomment the lines under the commented lines

# converting to gray scale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
#gray2 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

# removing noise
img2 = cv2.GaussianBlur(gray1,(3,3),0)
#img4 = cv2.GaussianBlur(gray2,(3,3),0)

# convoluting with Laplacian kernel
laplacian = cv2.Laplacian(img2,cv2.CV_64F)
#laplacian2 = cv2.Laplacian(img4,cv2.CV_64F)

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(img2,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

#plt.figure(figsize=(10,6))
#plt.subplot(1,1,1),plt.imshow(img4,cmap = 'gray')
#plt.title('Original'), plt.xticks([]), plt.yticks([])
#plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.show()

#plt.figure(figsize=(10,6))
#plt.subplot(1,1,1),plt.imshow(laplacian2,cmap = 'gray')
#plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
#plt.show()
```

### CANNY EDGE DETECTOR

<div style="text-align: justify"><font size="4">

The Process of Canny edge detection algorithm can be broken down to 4 different steps: <br><br>

    1. Apply Gaussian filter to smooth the image in order to remove the noise<br>
    2. Find the intensity gradients of the image<br>
    3. Apply non-maximum suppression to get rid of spurious response to edge detection<br>
    4. Apply hysteresis thresholding to determine potential edges<br><br><br><br>


<font size="5">Gaussian Filtering:</font><br><br>

Since all edge detection results are easily affected by image noise, it is essential to filter out the noise to prevent false detection caused by noise. To smooth the image, a Gaussian filter is applied to convolve with the image. It is important to understand that the selection of the size of the Gaussian kernel will affect the performance of the detector. The larger the size is, the lower the detectors sensitivity to noise. The kernel used to achieve this would be something like this:<br><br><br>

<img src="img/GaussianKernel.svg" align="center"><br><br><br>

<font size="5">Finding the intensity gradient of the image:</font><br><br>

An edge in an image may point in a variety of directions, so the Canny algorithm uses four filters to detect horizontal, vertical and diagonal edges in the blurred image. The edge detection operator returns a value for the first derivative in the horizontal direction (Gx) and the vertical direction (Gy). From this the edge gradient and direction can be determined:<br><br><br>

<img src="img/CannySobelFormula.png" align="center"><br><br>

Gradient direction is always perpendicular to edges. It is rounded to one of four angles representing vertical, horizontal and two diagonal directions.<br><br><br>

<font size="5">Non-maximum suppresion:</font><br><br>

Non-Maximum suppression is applied to "thin" the edge. After applying gradient calculation, the edge extracted from the gradient value is still quite blurred. With respect to criterion 3, there should only be one accurate response to the edge. Thus non-maximum suppression can help to suppress all the gradient values (by setting them to 0) except the local maxima, which indicate locations with the sharpest change of intensity value.<br><br><br>

<img src="img/nms.jpg" align="center"><br><br>

Point A is on the edge ( in vertical direction). Gradient direction is normal to the edge. Point B and C are in gradient directions. So point A is checked with point B and C to see if it forms a local maximum. If so, it is considered for next stage, otherwise, it is suppressed ( put to zero).<br><br>

In short, the result you get is a binary image with thin edges.<br><br><br>

<font size="5">Hysteresis Thresholding:</font><br><br> 

After application of non-maximum suppression, remaining edge pixels provide a more accurate representation of real edges in an image. However, some edge pixels remain that are caused by noise and color variation. In order to account for these spurious responses, it is essential to filter out edge pixels with a weak gradient value and preserve edge pixels with a high gradient value. <br><br>

For this, we need two threshold values, minVal and maxVal. Any edges with intensity gradient more than maxVal are sure to be edges and those below minVal are sure to be non-edges, so discarded. <br><br>

Those who lie between these two thresholds are classified edges or non-edges based on their connectivity. If they are connected to sure-edge pixels, they are considered to be part of edges. Otherwise, they are also discarded. The two threshold values are empirically determined and their definition will depend on the content of a given input image.<br><br><br>

<img src="img/hysteresis.jpg" align="center"><br><br>

The edge A is above the maxVal, so considered as sure-edge. Although edge C is below maxVal, it is connected to edge A, so that also considered as valid edge and we get that full curve. But edge B, although it is above minVal and is in same region as that of edge C, it is not connected to any sure-edge, so that is discarded. So it is very important that we have to select minVal and maxVal accordingly to get the correct result. This stage also removes small pixels noises on the assumption that edges are long lines.<br><br><br><br>

At the end of this process, our result will be a high accurate edge detection, and you probably may think that's a lot of work and yes it is... For the machine that calculates all that, nevertheless, OpenCV has all this process encapsulated into one single method. See the code example below where it is implemented.

</font></div>

```python
import cv2
import numpy as np
 
img = cv2.imread('img/mustang.jpg',cv2.IMREAD_GRAYSCALE)
img1 = cv2.imread('img/lambo.jpg',cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('img/sudoku.jpg',cv2.IMREAD_GRAYSCALE)
img3 = cv2.imread('img/hand.jpg',cv2.IMREAD_GRAYSCALE)
 
filtered = cv2.Canny(img,100,200)
filtered1 = cv2.Canny(img1,100,200)
filtered2 = cv2.Canny(img2,100,200)
filtered3 = cv2.Canny(img3,100,200)

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(img, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(filtered, cmap = 'gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(img1, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(filtered1, cmap = 'gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(img2, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(filtered2, cmap = 'gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(img3, cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.figure(figsize=(10,6))
plt.subplot(1,1,1),plt.imshow(filtered3, cmap = 'gray')
plt.title('Canny'), plt.xticks([]), plt.yticks([])
plt.show()
```
